```python
import pandas as pd
import numpy as np
import json
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

def import_sp_stop_words(jsonfile):
    with open(jsonfile, 'r') as file:
        jsondict = json.load(file)
    return jsondict['stop_words']

SPANISH_STOP_WORDS = import_sp_stop_words('stop_words_es.json')
```

```python

def tf_idf(word_data, max_df):
    ## Get stop words list from JSON FILE
    global SPANISH_STOP_WORDS
    vectorizer = TfidfVectorizer(sublinear_tf=True, min_df=max_df,
                                 stop_words=SPANISH_STOP_WORDS)
    x = vectorizer.fit_transform(word_data)
    features = vectorizer.get_feature_names()
    dense_matrix = x.toarray()
#     print(len(features))
#     print(dense_matrix)
#     print(features)
#     print(pd)
    return dense_matrix, features

def count_vectorizer(data, min_df):
    ## Get stop words list from JSON FILE
    global SPANISH_STOP_WORDS
    vectorizer = CountVectorizer(stop_words=SPANISH_STOP_WORDS, min_df=min_df)
    x = vectorizer.fit_transform(data)
    features = vectorizer.get_feature_names()
    dense_matrix = x.toarray()
#     print(len(features))
#     print(dense_matrix.shape)
#     print(features)
#     print(pd)
    return dense_matrix, features

def word_cloud(data, func=tf_idf, df = 0.8):
#     np.isnan(data)
    total_df = {}
    matrix, features = func(data, df)
#     print(features)
    data_frame = pd.DataFrame(matrix, columns=features)
    for f in features:
        count = np.sum(data_frame[f])
        total_df[f] = count
#     print(total_df)
    return {"word_cloud": total_df, "descriptor": 'text_data'}
    
if __name__ == '__main__':
    answer_data = {}
    reader = pd.read_csv('cuestionario_realizado.csv')
    c = 0
    for index, qa in reader.iterrows():
    #     print(qa)
        if answer_data.get(qa[0]) is None:
            answer_data[qa[0]] = []
        arr = answer_data.get(qa[0])
        arr.append(qa[1])
        answer_data[qa[0]] = arr
#     tests = np.arange(0.01, 0.1, 0.05)
#     for t in tests:
# #     print(tests)
#         print()
#         word_cloud_q1 = word_cloud(answer_data['Queremos mejorar el IGF Bolivia en sus próximas versiones, ¿tiene usted algún comentario o sugerencia?'], df=t)
#         word_cloud_q2 = word_cloud(answer_data['¿Qué temática hubiera querido incluir en el IGF Bolivia 2018?'], df=t)
    
#         df1 = pd.DataFrame(word_cloud_q1['word_cloud'], index=[0])
#         df2 = pd.DataFrame(word_cloud_q2['word_cloud'], index=[0])
#         df1.T.to_csv('word_cloud_q1T_max_df_{0}.csv'.format(t))
#         df2.T.to_csv('word_cloud_q2T_max_df_{0}.csv'.format(t))
        
#     print(df1)
#     df1 = pd.DataFrame.from_dict(word_cloud_q1['word_cloud'], index=[0])
#     d2 = pd.DataFrame.from_dict(word_cloud_q2['word_cloud'], index=[0])
#     print(df1)
#     print(df2)
    answer_data['Temáticas discutidas en los paneles'].append(4)
    
#     print(answer_data.keys())
#     print(answer_data)
    df = pd.DataFrame.from_dict(answer_data)
    print(df)
    df.to_csv('reporte_encuesta.csv')
    
#         print(splited)
```
